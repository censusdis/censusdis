# ACS Data Profile Demo

## Introduction

This notebook demonstrates how to load 
[US Census American Community Survey (ACS) 5-year Data Profiles](https://api.census.gov/data/2021/acs/acs5/profile.html).

## Imports and configuration

```python
# So we can run from within the censusdis project and find the packages we need.
import os
import sys

sys.path.append(
    os.path.join(os.path.abspath(os.path.join(os.path.curdir, os.path.pardir)))
)
```

```python
import censusdis.data as ced
import censusdis.maps as cem

from matplotlib.ticker import FuncFormatter
```

```python
# Set your API key here.
CENSUS_API_KEY = None
```

```python
YEAR = 2021
DATASET = "acs/acs5/profile"
```

```python
TOTAL_HOUSEHOLDS_VARIABLE = "DP02_0001E"
```

## Download the data

```python
gdf = ced.download(
    DATASET,
    YEAR,
    ['NAME', TOTAL_HOUSEHOLDS_VARIABLE],
    state="*",
    with_geometry=True
)
```

## Plot the data

```python
ax = cem.plot_us(
    gdf,
    TOTAL_HOUSEHOLDS_VARIABLE,
    figsize=(12, 6),
    legend=True,
    cmap="Blues",
    edgecolor="black",
    linewidth=0.25,
    vmin=0,
    vmax=15_000_000,
    legend_kwds={"format": FuncFormatter(lambda x, pos: f"{x:,.0f}")},
)

ax.axis("off")
ax.set_title("Total Households")

None
```
