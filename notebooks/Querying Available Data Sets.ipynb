# Querying Available Data Sets

The US Census publishes dozens of data sets per year. They have published over
1,300 data sets in all via their API. Some, like 
the *American Community Survey: 5-Year Estimates: Data Tables*
are widely known and used, but there are plenty of other interesting and useful
data sets you might want to load.

`censusdis` supports querying avaialble data sets for a given year or for all time.

We have not loaded all variables for all data sets for all years, but 
`censusdis` should be able to.
If you encounter one it cannot load, please file and issue at 
https://github.com/vengroff/censusdis/issues and we will look into it.

```python
# So we can run from within the censusdis project and find the packages we need.
import os
import sys

sys.path.append(
    os.path.join(os.path.abspath(os.path.join(os.path.curdir, os.path.pardir)))
)
```

## Imports and configuration

```python
import censusdis.data as ced
```

## Data sets available for a given year

```python
YEAR = 2020
```

```python
df_datasets_for_year = ced.variables.all_data_sets(year=YEAR)

df_datasets_for_year
```

## Data sets available for all years

```python
df_all_datasets = ced.variables.all_data_sets()

df_all_datasets
```

## List the groups in a dataset

```python
dataset = df_datasets_for_year.iloc[4]["DATASET"]
dataset
```

```python
df_groups = ced.variables.all_groups(dataset, YEAR)

df_groups
```

## List the variables in a group

```python
group = df_groups.iloc[-5]["GROUP"]
group
```

```python
df_variables = ced.variables.all_variables(dataset, YEAR, group)
df_variables
```

## Show the variables in a tree

```python
ced.variables.group_tree(dataset, YEAR, group)
```

## List all the leaves of the tree

```python
leaves = ced.variables.group_leaves(dataset, YEAR, group)
leaves
```

## Query the data

```python
# Note that these two are equivalent. Using `leaves_of_group=group`
# could have saved us needing to construct `leaves` with the
# separate call above.

df_download_variables = ced.download(dataset, YEAR, leaves, state="*")
df_download_leaves_of_group = ced.download(
    dataset, YEAR, leaves_of_group=group, state="*"
)

df_download_leaves_of_group.equals(df_download_variables)
```
