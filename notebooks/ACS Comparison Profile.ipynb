# ACS Comparison Profile Demo

## Introduction

This notebook demonstrates how to load and plot
[US Census American Community Survey (ACS) 5-year Comparison Profiles](https://api.census.gov/data/2021/acs/acs5/cprofile.html).

## Imports and configuration

```python
import censusdis.data as ced
import censusdis.maps as cem
from censusdis.states import ALL_STATES_AND_DC

from matplotlib.ticker import FuncFormatter
```

```python
# Set your API key here.
CENSUS_API_KEY = None
```

```python
DATASET_YEAR = 2021
DATASET = "acs/acs5/cprofile"
```

```python
PERCENT_POVERTY_VARIABLE = "CP03_2016_119E"
```

## Download the data

```python
gdf = ced.download(
    DATASET,
    DATASET_YEAR,
    ["NAME", PERCENT_POVERTY_VARIABLE],
    state=ALL_STATES_AND_DC,
    with_geometry=True,
)
```

## Plot the data

```python
ax = cem.plot_us(
    gdf,
    PERCENT_POVERTY_VARIABLE,
    figsize=(12, 6),
    legend=True,
    cmap="Reds",
    edgecolor="black",
    linewidth=0.25,
    vmin=0.0,
    vmax=20.0,
    legend_kwds={"format": FuncFormatter(lambda x, pos: f"{x:.0f}%")},
)

ax.axis("off")
ax.set_title("2012-2016 Change in Percent of Households Below the Poverty Line")

None
```
