# 2023 ACS1

On Septermber 12, 2024, the U.S. Census Bureau released the 
[ACS 1-Year Data](https://www.census.gov/data/developers/data-sets/acs-1year.html)
for 2023. This simple notebook demonstrates how `censusdis` was able to 
use the data immediately upon it's release.

```python
from datetime import datetime

datetime.now()
```

```python
import censusdis.data as ced
from censusdis.datasets import ACS1
from censusdis.states import ALL_STATES_AND_DC
```

```python
ced.variables.search(ACS1, 2023, pattern="median household income")
```

```python
ced.download(ACS1, 2023, ["NAME", "B19013A_001E"], state="*")
```
