# Blocks

This notebook demonstrates loading block level data. Note that very few data
sets and variables have data at the block level.

```python
import censusdis.data as ced
import censusdis.maps as cem
from censusdis.states import NJ
from censusdis.counties.new_jersey import HUDSON
from censusdis.datasets import DECENNIAL_PUBLIC_LAW_94_171
```

```python
gdf_blocks = ced.download(
    DECENNIAL_PUBLIC_LAW_94_171,
    2020,
    ["NAME"],
    state=NJ,
    county=HUDSON,
    block="*",
    with_geometry=True,
    remove_water=True,
)
```

```python
ax = cem.plot_map(
    gdf_blocks.boundary,
    figsize=(6, 10),
    color="black",
    linewidth=0.25,
)

ax.set_title("Census Blocks in Hudson County, NJ - 2020")
```
