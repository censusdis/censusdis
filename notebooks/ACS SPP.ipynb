Super-preliminary idea for Scott Pham for NICAR.

This shows how to use some of the new SPP detailed ethnicity data to
look at the Hmong population in Minnesota. Previously this would have been
rolled into Asian alone. But Minnesota has a large Hmong population.

Note that at the state level, only Minnesota and California have a large
enough Hmong population for it to be broken out in this data set.

```python
import censusdis.data as ced
from censusdis.datasets import ACS1_SPP
from censusdis.states import MN
```

```python
ced.variables.search(ACS1_SPP, 2022, group_name="S0201", pattern="POPGROUP")
```

```python
# See https://api.census.gov/data/2022/acs/acs1/spp/variables/POPGROUP.json

POPGROUP_ASIAN_ALONE = "012"
POPGROUP_HMONG_ALONE = "020"
```

```python
df_asian_alone = ced.download(
    ACS1_SPP,
    2022,
    ["NAME", "S0201_001E"],
    query_filter={"POPGROUP": POPGROUP_ASIAN_ALONE},
    state="*",
)

df_hmong_alone = ced.download(
    ACS1_SPP,
    2022,
    ["NAME", "S0201_001E"],
    query_filter={"POPGROUP": POPGROUP_HMONG_ALONE},
    state="*",
)
```
