```python
import censusdis.data as ced
import censusdis.maps as cem
from censusdis.states import PR
from censusdis.datasets import DECENNIAL_PUBLIC_LAW_94_171
```

```python
TOTAL_POPULATION_VARIABLE = "P1_001N"
```

```python
gdf_pr = ced.download(
    DECENNIAL_PUBLIC_LAW_94_171,
    2020,
    ["NAME", TOTAL_POPULATION_VARIABLE],
    state=PR,
    block="*",
    with_geometry=True,
)
```

```python
gdf_pr
```

```python
ax = cem.plot_map(
    gdf_pr, TOTAL_POPULATION_VARIABLE, cmap="Purples", legend=True, figsize=(10, 3)
)

_ = ax.set_title("Puerto Rico Population by Block - Decenial 2020")
```
